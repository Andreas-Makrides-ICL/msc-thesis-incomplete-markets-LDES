```python
import pandas as pd
import glob
import os

# Parameters
folder_path = r"C:\Users\user\Desktop\msc-thesis-incomplete-markets-LDES\models\Fully_incomplete_markets_ADMM\data_final\data_preparation_new\data_preparation_672\clustering\clustered_data"
output_capacity_file = r"C:\Users\user\Desktop\msc-thesis-incomplete-markets-LDES\models\Fully_incomplete_markets_ADMM\data_final\data_preparation_new\data_preparation_672\clustering\concatenated_capacity_factors_672_30yr_new.csv"
output_load_file = r"C:\Users\user\Desktop\msc-thesis-incomplete-markets-LDES\models\Fully_incomplete_markets_ADMM\data_final\data_preparation_new\data_preparation_672\clustering\concatenated_load_profiles_672_30yr_new.csv"
output_weights_file = r"C:\Users\user\Desktop\msc-thesis-incomplete-markets-LDES\models\Fully_incomplete_markets_ADMM\data_final\data_preparation_new\data_preparation_672\clustering\concatenated_weights_672_30yr_new.csv"

# Function to concatenate files matching a pattern
def concatenate_files(file_pattern, output_file):
    # Find all files matching the pattern
    files = glob.glob(os.path.join(folder_path, file_pattern))
    if not files:
        print(f"No files found for pattern {file_pattern}")
        return
    
    # Read and concatenate all files
    df_list = []
    for file in files:
        df = pd.read_csv(file)
        df_list.append(df)
        print(f"Read {file} with shape {df.shape}")
    
    # Concatenate vertically
    concatenated_df = pd.concat(df_list, ignore_index=True)
    
    # Save to output file
    concatenated_df.to_csv(os.path.join(folder_path, output_file), index=False)
    print(f"Saved {output_file} with shape {concatenated_df.shape}")

# Concatenate capacity factors files
concatenate_files('capacity_factors_gen_clustered_672_*.csv', output_capacity_file)
# Concatenate load profiles files
concatenate_files('load_profile_clustered_672_*.csv', output_load_file)
# Concatenate weights files
concatenate_files('weights_clustered_672_*.csv', output_weights_file)
```
